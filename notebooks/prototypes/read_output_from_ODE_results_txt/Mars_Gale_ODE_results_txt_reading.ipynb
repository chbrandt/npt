```python
file_input = 'ODE_Small_SearchResults.txt'
```

```python
import pandas as pd
# pd.set_option('display.max_colwidth', 200)
# pd.reset_option('max_colwidth')
```

```python
# parse column names
with open(file_input, 'r') as fp:
    fp.readline() # first line not of interest
    header_line = fp.readline()[1:] # discard first character ('#')
    column_names = [cn.strip().replace(' ','_') for cn in header_line.split(',')]
print('Column names:\n{!s}'.format(column_names))
```

```python
df = pd.read_csv(file_input, names=column_names, 
                 skip_blank_lines=True, skiprows=2, skipinitialspace=True)
df.head()
```

> **Notice the whitespaces ('\t') in those (string) columns. we have to trim them out!**

```python
_df_strings_ = df.select_dtypes(include='object')

df[_df_strings_.columns] = _df_strings_.apply(lambda s:s.str.strip())

del _df_strings_

df.head()
```
